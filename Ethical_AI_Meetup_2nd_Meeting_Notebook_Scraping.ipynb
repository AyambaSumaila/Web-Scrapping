## Introduction of Web-Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

Web-Scraping the Ebay Database

```python
# facebook.com/ - request
# html template - response
```

```python
# let define our 'url' and 'headers' which will bw used for the request
URL = "https://www.ebay.com/b/adidas-Ultraboost-Sneakers-for-Men/15709/bn_86579491b"
```

```python
HEADERS = ({"User-agent": "", "Accept-Language": "en-US,en;q=0.9"})
```

```python
# making a request to the url of Ebay
response = requests.get(URL, headers=HEADERS)
```

```python
response.status_code # we get 200 status to verify the request was successful
```

```python
response_content = response.content # this give the raw binary content of our response
```

```python
type(response_content)
```

```python
# lets view the contents
response_content[:4000]
```

What we need to do is to parse the HTML content of the response which is initially in binary

```python
# we need to get the parsed HTMl contents suing the beautiful soup package
html_template = BeautifulSoup(response.content, "html.parser")
```

```python
html_template
```

We continue on our next meeting 😊

```python
links = html_template.find_all('a', attrs="s-item__link")
```

```python
link = links[0].get('href')
```

```python
new_webpage = requests.get(link, headers=HEADERS)
```

```python
new_webpage.content
```

```python
soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
soup
```

```python
soup.find('h1', attrs={"class": "x-item-title__mainTitle"}).text.strip()
```
